# Respuesta al impulso

## Sweep Logaritmico

La ecuación que describe el comportamiento de un sine sweep logaritmico es la siguiente

\begin{gather*}
x(t) = \sin(K(e^{t/L}-1)) 
\end{gather*}
donde:
\begin{gather*}
K = \dfrac{2\pi f_1 T}{\ln\left(\dfrac{f_2}{f_1}\right)} & L = \dfrac{T}{\ln\left(\dfrac{f_2}{f_1}\right)}  
\end{gather*}

siendo $f_1$ la frecuencia inicial, $f_2$ la frecuencia final y $T$ la duración de la señal

```python
sr = 96000
T = 25
f1 = 42
f2 = 18000
sweep_time, sweep_magnitude = sys.sweep(T, f1, f2, sr) # sweep signal
```

## Filtro Inverso

El filtro inverso es calculado a partir del escalamiento en amplitud $m(t)$ del sine sweep invertido en tiempo $x(-t)$. Lo cual resulta en un sweep con decrecimiento exponencial. 
\begin{gather*} 
f(t) = m(t)x(-t)
\end{gather*}

para conocer la velocidad a la que aumenta la frecuencia se realizó la derivación del argumento del seno utilizado, 

\begin{gather*} 
\theta(t) = K(e^{t/L}-1) \Rightarrow \dfrac{d \theta(t)}{dt} = \dfrac{K}{L} e^{t/L}
\end{gather*}

simplificando:
\begin{gather*} 
\dfrac{d \theta(t)}{dt} = 2\pi f_1 e^{t/L} = \omega_1 e^{t/L}
\end{gather*}
Para deshacerse de la dependencia del termino $\omega_1$, de forma tal que la amplitud máxima de dicha modulación sea 1, se define a la función $m(t)$ como:

\begin{gather*} 
m(t) = \dfrac{\omega_1}{\omega_1 e^{t/L}} \Rightarrow m(t) = \dfrac{1}{e^{t/L}} = e^{-t/L}
\end{gather*}

Obteniendose así el filtro inverso como,
\begin{gather*} 
f(t) = e^{-t/L}x(-t)
\end{gather*}

```python
inv_filter_time, inv_filter_magnitude = sys.inverse_filter((sweep_time, sweep_magnitude), T, f1, f2) # inverse filter
```

```python
# load audio files recording
files = ['1_sweep_response_L.wav', '1_sweep_response_R.wav', '2_sweep_response_L.wav', '2_sweep_response_R.wav']
recordings = [] 
ir_responses = []
freq_responses = []
# save each recording on array compose of magnitude and sample rate
for file in files:
    recordings.append(sf.read(os.path.join(DIRNAME, 'audios', 'processed', file)))

# save impulse responses on list, each position is np.ndarray composed of time and magnitude data
for rec in range(0, len(recordings)):
    ir = sys.impulse_response(recordings[rec][0], inv_filter_magnitude, sr)
    ir = [ir[0], np.real(ir[1])] # get only the real component
    ir_responses.append(ir)
    
# ir_time, ir_magnitude = sys.impulse_response(mag_inv_filter, rec_sweep, sr)
# ir_fft = sys.fft(ir_magnitude)
# ir_fft_freq = np.linspace(0, sr, len(ir_fft))
# sys.grapher([[ir_fft_freq, ir_fft]], dimension=(1,1), semilog=True, xlim=[20, 20000], xlabel=[])

for ir in range(0, len(ir_responses)):
    ir_fft = sys.fft(ir_responses[ir][1])
    frequencies = np.linspace(0, sr/2, len(ir_fft))
    freq_responses.append([frequencies, ir_fft])
```

Para calcular las frecuencias percibidas $f_p$ de una señal cualquiera de frecuencia $f$, la cual es muestreada a una frecuencia $f_s$, se utiliza la siguiente formula:

\begin{equation*}
f_p = \left| f-f_s\cdot NINT\left(\dfrac{f}{f_s}\right) \right|
\end{equation*}

Donde $NINT$ es la función que entrega el entero mas cercano realizando un redondeo hacia arriba. Por ej, para 10.50 redondea a 11 y 10.49 redondea a 10.
