# Transformada de Fourier

La Transformada de Fourier nos permite describir el espectro en frecuencia de una señal. Usando la transformada de Fourier, tanto las señales periódicas como las no periódicas se pueden transformar del dominio del tiempo al dominio de la frecuencia.

\begin{equation}
X(j\omega) = \frac{1}{2\pi} \int_{-\infty}^{\infty} \! x(t) \mathrm{e}^{-j\omega t}\, dt
\end{equation}


Existen muchas maneras de interpretar los resultados y las propiedades de esta transformada. Usualmente, el enfoque geométrico resulta el más amigable de asimilar, dado que se puede visualizar con [animaciones intuitivas](https://www.youtube.com/watch?v=h4PTucW3Rm0&ab_channel=3Blue1BrownEspa%C3%B1ol). 

### Aplicaciones
Algunos pocos ejemplos de los muchos en los cuales se emplea la transformada de fourier son:

- Relacionado a las vibraciones, se utiliza para determinar las frecuencias dominantes en una señal de vibración. Cuando la frecuencia dominante de una señal se corresponde con la frecuencia natural de una estructura, las vibraciones que se producen pueden amplificarse debido a la resonancia. Esto puede suceder hasta tal punto que una estructura puede colapsar.

- En la acústica de recintos, se emplean gráficas en el dominio de la frecuencia (amplitud vs frecuencia) para obtener descriptores acústicos y realizar análisis y diseños

- En el procesamiento de señales de habla el dominio frecuencial al cual se accede por la transformada de Fourier es altamente utilizado en los modelos de lenguaje, para el estudio de los fonemas y sus caracteristicas.

# Interpretación sobre una señal temporal

Usualmente usamos la transformada de Fourier para pasar del dominio __temporal__ al dominio __frecuencial__, lo que comúnmente denominamos _espectro_.

Vamos a ver un ejemplo de una señal __temporal__, en este caso, una parte de una cación de la banda _Justice_ llamada  _Safe and Sound_ 

```python
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio

audio, sr = sf.read('justice.wav')
time = np.linspace(0, len(audio)/sr, len(audio))
plt.plot(time, audio)
plt.subplots_adjust(right=2.5, top=1.5)
Audio(audio, rate=sr)
```

A simple vista pareciera ser una forma de onda compuesta por distintas frecuencias (no luce como una onda senoidal pura). Esto se deduce de observar zonas que parecen responder a períodos diferentes. 

Por simple inspección visual, si contamos los picos, podemos ver que en un cuarto de segundos hay 15 picos. Es decir que en un segundo, tendríamos 60 picos. En señales senoidales, 2 picos corresponden a 1 período. Es decir que tengo 30 períodos en 1 segundo (aproximadamente), lo que se corresponde con una onda senoidal de 30 Hz (frecuencia bastante baja). 

Para hacer este análisis un poco más certero, podemos usar la Transformada de Fourier. En términos prácticos, Fourier nos dice que cualquier función puede ser representada como una suma de señales senoidales. Es decir, podemos descomponer cualquier forma de onda como una suma de ondas senoidales. 

Esto se ilustra en la imagen debajo, donde se muestra la descomposición de una señal cuadrada.

Esto es lo que logramos con la transformada de Fourier: tomar una señal temporal y obtener información a cerca de todas las señales senoidales necesarias para conformar esa señal temporal.

## En el dominio discreto

Para secuencias temporales de valores equiespaciados la Transformada de Fourier Discreta (DFT) se define como:

\begin{equation}
X[k] = \sum_{n=0}^{N-1} \! x[n] \mathrm{e}^{-j2\pi kn/N} dt
\end{equation}

donde:

* N = numero de muestras
* n = muestra actual
* x[n] = valor de la señal para n
* k = valor de frecuencia (0 Hz hasta N-1 Hz)
* X[k] = resultado de la DFT (amplitud y fase)

nota: recordar que el producto punto se puede definir como:

\begin{equation}
a \cdot b = \sum_{n=1}^{N} a_{i} b_{i}
\end{equation}

Si bien esta operación matemática se puede llevar a código muy facilmente, su implementación directa es subóptima en términos de eficiencia computacional. Afortunadamente, [existen implementaciónes recursivas](https://en.wikipedia.org/wiki/Cooley%E2%80%93Tukey_FFT_algorithm) de este algoritmo que logran reducir considerablemente la complejidad computacional.

De la documentación de Numpy.fft...

*FFT (Fast Fourier Transform) refers to a way the discrete Fourier Transform (DFT) can be calculated efficiently, by using symmetries in the calculated terms. The symmetry is highest when n is a power of 2, and the transform is therefore most efficient for these sizes.
(Fourier analysis is fundamentally a method for expressing a function as a sum of periodic components, and for recovering the function from those components. When both the function and its Fourier transform are replaced with discretized counterparts, it is called the discrete Fourier transform (DFT). The DFT has become a mainstay of numerical computing in part because of a very fast algorithm for computing it, called the Fast Fourier Transform (FFT), which was known to Gauss (1805) and was brought to light in its current form by Cooley and Tukey. 
Because the discrete Fourier transform separates its input into components that contribute at discrete frequencies, it has a great number of applications in digital signal processing, e.g., for filtering, and in this context the discretized input to the transform is customarily referred to as a signal, which exists in the time domain. The output is called a spectrum or transform and exists in the frequency domain.*

# Transformada de Fourier --> Numpy.fft.fft 

<img src="imagenes/numpyfft.png" alt="Drawing"/>

```python
fft_raw = np.fft.fft(sum_sin)
```

En la celda anterior, se obtuvo la transformada de Fourier de la suma de senoidales. Inspeccionemos algunos valores de este resultado:

```python
print(fft_raw)
```

Hay varias cuestiones que podemos notar de esta simple inspección inicial. En primer lugar, vemos que el tipo de dato que compone este array son números complejos. Luego, se puede ver que el valor __n__ pareciera tener la misma parte real que el elementeo __-n__ y la parte imaginaria es inversa. En otras palabras, son los respectivos valores __complejos conjugados__.

Para no usar indices negativos, podemos decir que el valor __n__ es el complejo conjugado de __N-n__ para n distinto de cero, donde N es el número de elementos. 

Viendo esto, podemos interpretar como obtenemos la salida esperada de la función __np.fft.fft__. Por definición tenemos un espectro simétrico. Lo usual es quedarse con la parte positiva del mismo, que va desde la frecuencia correspondiente a _0 Hz_ en adelante. Para ello, deberiamos tomar las muestras desde _0_ hasta _N//2_ siendo N el número de muestras de la señal.

<img src="imagenes/espectros.png" alt="Drawing" style="height : 400pt"/>

```python
fft = fft_raw[:len(fft_raw)//2]
print(fft)
```

Lo siguiente es pasar de valores complejos a reales. Lo usual es analizar por un lado la magnitud de la transformada, y por otro lado la fase. Siguiendo con el enfoque anterior, podemos pensar a la magnitud como la __Amplitud__ de una senoidal a una determinada frecuencia (proporcional a la energía que contiene) y a la fase como el punto de offset de dicha senoidal. Veamos primero la magnitud, para lo cual tomamos el valor absoluto del numero complejo. Además, se debe escalar el valor de amplitud por el largo de la señal, debido a la proyección sobre las frecuencias que se calcula durante la implementación de la fft.

```python
fft_mag = abs(fft) / len(fft)
```

Para poder interpretar el espectro necesitamos graficarlo en función de los correspondientes valores de frecuencia (eje X). Cada punto discreto de la transformada se corresponde con un determinado valor de frecuencia. La resolución frecuencial viene dada por:

\begin{equation}
\Delta f = \frac{f_{s}}{N}
\end{equation}

Otra manera de obtener este eje es sabiendo que la frecuencia minima que se representa es _0 Hz_, y por __teorema de Nyquist__, la máxima es _sr/2_. Entonces, para la parte positiva del espectro, se puede definir el eje de frecuencias como:

```python
freqs = np.linspace(0, fs/2, len(fft))
```

Con esto podemos hacer un primer gráfico para ver que forma tiene esta señal

```python
plt.plot(freqs, fft_mag)
```

hagamos un "zoom" en frecuencia 

```python
plt.plot(freqs, fft_mag)
plt.xlim(0, 600)
```

Rapidamente, podemos comprobar que la lectura del espectro nos indica que la señal es la suma de dos componentes, una de 250 Hz a una amplitud de 5, y otra de 500 Hz a una amplitud de 10.

Todavia podemos mejorar aún mas esta representación, teniendo en cuenta algunos conceptos que ya vimos con respecto a gráficos en frecuencia. Uno de ellos es la escala del eje X, el cual sabemos que es mas representativo al verlo en escala logaritmica. Además, podemos indicar puntos en el eje de frecuencia que sean comúnmente analizados, como centros nominales de bandas de octava.

Respecto a la amplitud, por la forma en que nuestro sistema auditivo percibe los cambios de niveles de presión sonora, es mas representativo mostrar las magnitudes en __dB__. Para ello, podemos referenciar los valores al valor máximo de la señal, llevándolo al punto de 0 dB. Esto equivale a primero normalizar la señal, y luego pasarla a dB. 

```python
fft_mag_norm = fft_mag / np.max(abs(fft_mag))
eps = np.finfo(float).eps
fft_mag_db = 20*np.log10(fft_mag_norm + eps)

plt.figure(figsize=(15,5))
plt.semilogx(freqs, fft_mag_db)
ticks = [63, 125, 250, 500, 1000, 2000, 4000, 8000]
plt.xticks([t for t in ticks], [f'{t} Hz' for t in ticks])
plt.xlim(50, 22000)
plt.ylim(-30, 5)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.grid()
plt.show()
```

vemos una diferencia de 3 dB en energía, lo que se corresponde con una relación 2 a 1. 

De igual forma, podemos ver la información de fase

```python
fft_pha = np.angle(fft)
plt.figure(figsize=(15,5))
plt.semilogx(freqs, fft_pha)
ticks = [63, 125, 250, 500, 1000, 2000, 4000, 8000]
plt.xticks([t for t in ticks], [f'{t} Hz' for t in ticks])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [rad]')
plt.grid()
plt.show()
```

# Para un audio real

```python
import soundfile as sf
from IPython.display import Audio
```

```python
audio, fs = sf.read('justice.wav')
Audio(audio, rate=fs)
```

```python
fft_raw = np.fft.fft(audio)
fft = fft_raw[:len(fft_raw)//2]
fft_mag = abs(fft) / len(fft)
freqs = np.linspace(0, fs/2, len(fft))
```

```python
fft_mag_norm = fft_mag / np.max(abs(fft_mag))
eps = np.finfo(float).eps
fft_mag_db = 20*np.log10(fft_mag_norm + eps)

plt.figure(figsize=(15,5))
plt.semilogx(freqs, fft_mag_db)
ticks = [63, 125, 250, 500, 1000, 2000, 4000, 8000]
plt.xticks([t for t in ticks], [f'{t} Hz' for t in ticks])
plt.xlim(20, 22000)
plt.ylim(-80, 5)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.grid()
plt.show()
```
